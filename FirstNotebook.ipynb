```python
# Cell 1 - Environment setup
import os
os.environ['TF_DISABLE_METAL'] = '1'
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import tensorflow as tf
import numpy as np
import random as rn
import shutil

# Clear old logs
if os.path.exists('logs'):
    shutil.rmtree('logs')
    print("Old logs cleared ✅")

tf.keras.backend.clear_session()
np.random.seed(0)
tf.random.set_seed(0)
rn.seed(0)

print("TF Version:", tf.__version__)
print("Eager mode:", tf.executing_eagerly())  # Keep True ✅

# Cell 2 - Load data
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 784) / 255.0
X_test = X_test.reshape(X_test.shape[0], 784) / 255.0
Y_train = tf.keras.utils.to_categorical(y_train, 10)
Y_test = tf.keras.utils.to_categorical(y_test, 10)

# Cell 3 - Build model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import History, TensorBoard
from tensorflow.keras.models import Model

input_layer = Input(shape=(784,))
layer1 = Dense(50,
    activation='sigmoid',
    name='dense_1',
    kernel_initializer=tf.keras.initializers.glorot_normal(seed=0))(input_layer)
output = Dense(10,
    activation='softmax',
    name='dense_2',
    kernel_initializer=tf.keras.initializers.glorot_normal(seed=0))(layer1)

model = Model(inputs=input_layer, outputs=output)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Cell 4 - Train
history = History()
tf_tensor = TensorBoard(
    log_dir='logs',
    histogram_freq=1,
    write_graph=True,
    write_images=True,
    update_freq='epoch'
)

model.fit(
    X_train, Y_train,
    epochs=3,
    validation_data=(X_test, Y_test),
    batch_size=512,
    callbacks=[history, tf_tensor]
)

# Cell 5 - Save graph manually after training
print("\nSaving graph to TensorBoard...")

# Create a concrete function for graph tracing
@tf.function
def serving_fn(x):
    return model(x)

# Trace and write graph
writer = tf.summary.create_file_writer('logs/train')
tf.summary.trace_on(graph=True, profiler=False)

# Run one forward pass to trace graph
serving_fn(tf.zeros([1, 784]))

with writer.as_default():
    tf.summary.trace_export(
        name="model_trace",
        step=0
    )
writer.flush()
print("Graph saved successfully ✅")
```
